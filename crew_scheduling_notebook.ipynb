# Airline Crew Scheduling using Backtracking
This notebook implements the assignment requirements.

```

flights = [('F1', 9, 11), ('F2', 10, 12), ('F3', 13, 15)]
crew_members = ['C1', 'C2', 'C3']

def is_valid(flight, assigned):
    for f in assigned:
        if not (flight[2] + 1 <= f[1] or f[2] + 1 <= flight[1]):
            return False
    return True

solution = {c: [] for c in crew_members}
best = []

def backtrack(i, sol):
    global best
    if i == len(flights):
        best = [list(sol[c]) for c in sol]
        return True
    for c in crew_members:
        if is_valid(flights[i], sol[c]):
            sol[c].append(flights[i])
            if backtrack(i+1, sol):
                return True
            sol[c].pop()
    return False

backtrack(0, solution)
solution
```
